The Python Imaging Library (PIL) is an open-source python module for image handling and processing. Pillow is a friendly fork of PIL. Since the original version of PIL is not maintained anymore, Pillow represents the currently most used version of PIL.

Installation (Ubuntu)
* $ pip install pillow

Resources
* Documentation: https://pillow.readthedocs.io/en/4.0.x/index.html
* Tutorial: https://pillow.readthedocs.io/en/latest/handbook/tutorial.html


### Importing

```python
# All functionalities which we will use are contained in the Image class
from PIL import Image

print(Image.__version__) # Pillow version
```

### Input / Output

```python
# Loading an image file
img = Image.open("Data/robotic_kitten.png")

# Show the datatype of the loaded image
print("Datatype:", type(img))

# Saving a file to drive
img.save("Data/robotic_kitten_copy.png")

# Veryfy saving
import os
print("Saved 'robotic_kitten_copy.png':", "robotic_kitten_copy.png" in os.listdir("Data"))

# Remove copy from disk
os.remove("Data/robotic_kitten_copy.png")
```

```python
# Show an image in separate window
img.show()
```

```python
# PIL images can be easily displayed in jupyter notebook
from IPython.display import display 
display(img)
```

```python
# Create numpy array from PIL image
import numpy as np

array = np.array(img)

print(type(img))
print(type(array))
```

```python
# Remove transparency channel from RGBA image
print(array.shape)

array_rgb = array[:, :, :3]

print(array_rgb.shape)
```

### Image meta data

```python
# Print image file format
print("Image format:", img.format)

# Print image size in pixel along each axis
print("Image size:", img.size)

# Print the channel mode of an image
print("Image mode:", img.mode)
```

### Transformations

```python
# Resize an image
resized = img.resize((256, 128))
display(resized)
```

```python
# Rotate an image
rotated = img.rotate(180) # Parameter indicates the rotation in degrees counter-clockwise
display(rotated)
```

```python
# Flip an image
flipped = img.transpose(Image.FLIP_LEFT_RIGHT)
display(flipped)
```

### Cropping

```python
# Cropping an image by defining a region to extract
bbox = (250, 0, 600, 400)  # left, upper, right, lower
roi = img.crop(bbox)

print("Original size in pixels:", img.size)
print("Cropped size in pixels:", roi.size)

display(roi)
```
